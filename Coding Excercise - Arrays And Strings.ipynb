```python
public class Solution {
    public int RemoveDuplicates(int[] nums) {
        List<int> occuringNums = new List<int>();

        int expectedNums = 0;
        for (int i = nums.Length-1; i >= 0; i--){
            int currentNum = nums[i];
            if (!occuringNums.Contains(currentNum)){
                occuringNums.Add(currentNum);
                expectedNums++;
                Console.WriteLine($"i: {i}, currentNum: {currentNum}. expectedNums: {expectedNums}");
            }
        }

        occuringNums.Sort();
        for (int i = 0; i < occuringNums.Count; i++){
            nums[i] = occuringNums[i];
        }

        return expectedNums;
    }
}
```

169. Majority Element

Given an array nums of size n, return the majority element.

The majority element is the element that appears more than ⌊n / 2⌋ times. You may assume that the majority element always exists in the array.



Example 1:

Input: nums = [3,2,3]
Output: 3
Example 2:

Input: nums = [2,2,1,1,1,2,2]
Output: 2


Constraints:

n == nums.length
1 <= n <= 5 * 104
-109 <= nums[i] <= 109


Follow-up: Could you solve the problem in linear time and in O(1) space?

```python
#Solution 1

public class Solution {
    public int MajorityElement(int[] nums) {
        Dictionary<int, int> numCount = new Dictionary<int,int>();

        foreach(int num in nums){
            try{
                numCount[num]++;
            }
            catch{
                numCount[num]=1;
            }
        }

        return numCount.OrderByDescending(x => x.Value).First().Key;
    }
}
```

```python
public class Solution {
    public void Rotate(int[] nums, int k) {
        int[] rotated = new int[nums.Length];
        for (int i = 0; i < nums.Length; i++){
            int num = nums[i];
            int newIndex = i+k;
            if(newIndex>=nums.Length){
                newIndex = newIndex % nums.Length;
            }

            rotated[newIndex]=num;
        }

        for (int i = 0; i < nums.Length; i++){
            nums[i] = rotated[i];
        }
    }
}
```

Find and return the maximum profit you can achieve.



Example 1:

Input: prices = [7,1,5,3,6,4]
Output: 7
Explanation: Buy on day 2 (price = 1) and sell on day 3 (price = 5), profit = 5-1 = 4.
Then buy on day 4 (price = 3) and sell on day 5 (price = 6), profit = 6-3 = 3.
Total profit is 4 + 3 = 7.
Example 2:

Input: prices = [1,2,3,4,5]
Output: 4
Explanation: Buy on day 1 (price = 1) and sell on day 5 (price = 5), profit = 5-1 = 4.
Total profit is 4.
Example 3:

Input: prices = [7,6,4,3,1]
Output: 0
Explanation: There is no way to make a positive profit, so we never buy the stock to achieve the maximum profit of 0.


Constraints:

1 <= prices.length <= 3 * 104
0 <= prices[i] <= 104

```python
public class Solution {
    public int MaxProfit(int[] prices) {
        bool isBought = false;
        int currentStockWorth = 0;
        int totalProfit = 0;
        for (int i = 0 ; i < prices.Length; i++){
            int todaysPrice = prices[i];
            int tomorrowPrice = 0;

            if(i<prices.Length-1){
                tomorrowPrice = prices[i+1];
            }


            if(!isBought && todaysPrice<tomorrowPrice){
                currentStockWorth = todaysPrice;
                isBought = true;
                //Console.WriteLine($"Buy: i: {i}. todaysPrice: {todaysPrice}. tomorrowPrice: {tomorrowPrice}.");
            }
            else if (isBought && todaysPrice>tomorrowPrice){
                int profit = todaysPrice-currentStockWorth;
                totalProfit += profit;
                currentStockWorth = 0;
                isBought = false;
                //Console.WriteLine($"Sell: i: {i}. todaysPrice: {todaysPrice}. tomorrowPrice: {tomorrowPrice}. currentStockWorth: {currentStockWorth}. profit: {profit}. totalProfit: {totalProfit}");
            }

        }
        return totalProfit;
    }
}
```

Return true if you can reach the last index, or false otherwise.



Example 1:

Input: nums = [2,3,1,1,4]
Output: true
Explanation: Jump 1 step from index 0 to 1, then 3 steps to the last index.
Example 2:

Input: nums = [3,2,1,0,4]

Explanation: You will always arrive at index 3 no matter what. Its maximum jump length is 0, which makes it impossible to reach the last index.


Constraints:

1 <= nums.length <= 104
0 <= nums[i] <= 105

```python
public class Solution {
    public bool CanJump(int[] nums) {
        bool canReachEnd = false;
        bool stepsLeft = true;
        int currentJumpLen = nums[0];
        int currentIndex = 0;

        List<int> selectedIndexes = new List<int>();

        if (nums.Length==1){
            return true;
        }

        // Validate that the end is even reachable:
        if(nums.Length>100){
            Console.WriteLine($"List Length: {nums.Length}.");
            bool isReachable = false;
            for (int j=nums.Length-1; j>=0; j--){
                Console.WriteLine($"j: {j}. val: {nums[j]}. Total: {j+nums[j]}.");
                if(j!=nums.Length-1 && j+nums[j]>=nums.Length-1){
                    isReachable = true;
                    Console.WriteLine($"isReachable: {isReachable}.");
                    break;
                }
                }
                if (!isReachable){return false;}
            }


        while (stepsLeft && !canReachEnd){

            if(currentIndex+currentJumpLen >= nums.Length-1){
                canReachEnd = true;
                break;
            }
            if(currentJumpLen==0){
                canReachEnd = false;
                break;
            }


            int[] jumpRange = nums.Skip(currentIndex+1).Take(currentJumpLen).ToArray();
            int maxIndex = Array.LastIndexOf(jumpRange, jumpRange.Max())+1;



            currentIndex += maxIndex;
            currentJumpLen = jumpRange.Max();

            if(currentJumpLen!=0 && !selectedIndexes.Contains(currentIndex)){
                selectedIndexes.Add(currentIndex);
            }

            if(currentJumpLen==0){
                if(selectedIndexes.Count == 0){
                    canReachEnd = false;
                    break;
                }

                int indexToReset = selectedIndexes.Max();
                nums[indexToReset] = 0;
                selectedIndexes.Remove(indexToReset);
                currentJumpLen = nums[0];
                currentIndex = 0;
                Console.WriteLine($"No Steps Left. Will remove index: {indexToReset} and reattempt flow");
                Console.WriteLine($"selected steps: {"[" + string.Join(",", selectedIndexes) + "]"}");
                Console.WriteLine($"nums after change: {"[" + string.Join(",", nums) + "]"}");
            }

        }

        return canReachEnd;
    }
}
```

45. Jump Game II

You are given a 0-indexed array of integers nums of length n. You are initially positioned at nums[0].

Each element nums[i] represents the maximum length of a forward jump from index i. In other words, if you are at nums[i], you can jump to any nums[i + j] where:

0 <= j <= nums[i] and
i + j < n
Return the minimum number of jumps to reach nums[n - 1]. The test cases are generated such that you can reach nums[n - 1].



Example 1:

Input: nums = [2,3,1,1,4]
Output: 2
Explanation: The minimum number of jumps to reach the last index is 2. Jump 1 step from index 0 to 1, then 3 steps to the last index.
Example 2:

Input: nums = [2,3,0,1,4]
Output: 2


Constraints:

1 <= nums.length <= 104
0 <= nums[i] <= 1000
It's guaranteed that you can reach nums[n - 1].

```python
# Solution 1
public class Solution {
    public int Jump(int[] nums) {

        if (nums[0]==0 || nums.Length == 1){
            return 0;
        }
        else if (nums[0]>=nums.Length-1){
            return 1;
        }

        List<int> numOfAttempts = new List<int>();
        List<int> validSteps = new List<int>();

        bool resetGame = false;
        int currentIndex = 0;
        int currentJumpRange = nums[currentIndex];

        int stepCount = 0;

        while(1==1){
            if(resetGame){
                if(validSteps.Count==0){
                    Console.WriteLine($"break");
                    break;
                }

                int maxLastIndex = validSteps.Max();
                nums[maxLastIndex] = 0;
                currentIndex = 0;
                currentJumpRange = nums[currentIndex];
                Console.WriteLine($"Reset Round: maxLastIndex: {maxLastIndex}. validSteps: {"[" + string.Join(",", validSteps) + "]"}");
                Console.WriteLine($"nums after change: {"[" + string.Join(",", nums) + "]"}");
                validSteps.Remove(maxLastIndex);
                stepCount = 0;
            }

            stepCount++;

            int[] slice = nums.Skip(currentIndex+1).Take(currentJumpRange).ToArray();
            int maxJumpValue = slice.Max();

            int maxIndex = Array.LastIndexOf(slice,maxJumpValue)+1;
            int maxReachAbility = currentIndex+currentJumpRange;

            Console.WriteLine($"stepCount: {stepCount}. currentIndex: {currentIndex}. currentJumpRange: {currentJumpRange}. maxJumpValue: {maxJumpValue}. maxIndex: {maxIndex}. Final Index: {currentIndex+maxIndex}. maxReachAbility: {maxReachAbility}");
            currentJumpRange = maxJumpValue;

            currentIndex+=maxIndex;

            if(!validSteps.Contains(currentIndex) && maxJumpValue!=0){
                Console.WriteLine($"Add index: currentIndex {currentIndex}.");
                validSteps.Add(currentIndex);
            }

            if(currentJumpRange==0 && currentIndex<nums.Length-1){
                resetGame = true;
                Console.WriteLine($"Dead End!: stepCount {stepCount}. currentIndex: {currentIndex}. nums.Length: {nums.Length}.");
            }





            if(currentIndex>=nums.Length-1 || maxReachAbility>=nums.Length-1){
                numOfAttempts.Add(stepCount);
                Console.WriteLine($"Success!: stepCount {stepCount}. currentIndex: {currentIndex}. nums.Length: {nums.Length}. maxReachAbility: {maxReachAbility}");
                resetGame = true;

                if (stepCount<=2){
                    return stepCount;
                }
            }


        }

        return numOfAttempts.Min();
    }
}
```

135. Candy

There are n children standing in a line. Each child is assigned a rating value given in the integer array ratings.

You are giving candies to these children subjected to the following requirements:

Each child must have at least one candy.
Children with a higher rating get more candies than their neighbors.
Return the minimum number of candies you need to have to distribute the candies to the children.



Example 1:

Input: ratings = [1,0,2]
Output: 5
Explanation: You can allocate to the first, second and third child with 2, 1, 2 candies respectively.
Example 2:

Input: ratings = [1,2,2]
Output: 4
Explanation: You can allocate to the first, second and third child with 1, 2, 1 candies respectively.
The third child gets 1 candy because it satisfies the above two conditions.


Constraints:

n == ratings.length
1 <= n <= 2 * 104
0 <= ratings[i] <= 2 * 104

```python
public class Solution {
    public int Candy(int[] ratings) {

        Dictionary<int,int> kidsAndScores = new Dictionary<int,int>();
        for(int i = 0; i<ratings.Length; i++){
            kidsAndScores[i]=1;
        }

        List<int> indexesToIgnore = new List<int>();

        int ascCount = 0;
        int dscCount = 0;

        int index = 0;
        while(index<ratings.Length){
            bool anyChange = false;

            for(int i = 0; i<ratings.Length; i++){

                if(!indexesToIgnore.Contains(i)){
                    int currentChildRating = ratings[i];
                    int currentChildCandy = kidsAndScores[i];
                    int previousChildRating = currentChildRating;
                    int previousChildCandy = currentChildCandy;

                    if(i-1!=-1){
                        previousChildRating = ratings[i-1];
                        previousChildCandy = kidsAndScores[i-1];
                    }
                    int nextChildRating = currentChildRating;
                    int nextChildCandy = currentChildCandy;
                    if(i+1!=ratings.Length){
                        nextChildRating = ratings[i+1];
                        nextChildCandy = kidsAndScores[i+1];
                    }

                    if(previousChildRating<currentChildRating && previousChildCandy>=currentChildCandy){
                        kidsAndScores[i] = previousChildCandy+1;
                        anyChange = true;
                    }
                    if(nextChildRating<currentChildRating && nextChildCandy>=currentChildCandy){
                        kidsAndScores[i] = nextChildCandy+1;
                        anyChange = true;
                    }

                    if(previousChildRating>currentChildRating && nextChildRating>currentChildRating){
                        indexesToIgnore.Add(i);
                        Console.WriteLine($"Ignore index: {i}. previousChildRating: {previousChildRating}. currentChildRating: {currentChildRating}. nextChildRating: {nextChildRating}.");
                    }

                    if(index == 0 && previousChildRating<currentChildRating){
                        ascCount++;
                    }
                    else if(index == 0 && previousChildRating>currentChildRating){
                        dscCount++;
                    }

                }


            }
            if (!anyChange){
                break;}

            if(dscCount==ratings.Length-1 || ascCount==ratings.Length-1){
                int comulativeCount = ratings.Length * (ratings.Length +1) / 2;
                return comulativeCount;
            }
            index++;
        }

        foreach(var kidsAndScore in kidsAndScores){
            Console.WriteLine($"Key: {kidsAndScore.Key}. Val: {kidsAndScore.Value}.");
        }
        return kidsAndScores.Values.Sum();
    }
}
```

68. Text Justification


Given an array of strings words and a width maxWidth, format the text such that each line has exactly maxWidth characters and is fully (left and right) justified.

You should pack your words in a greedy approach; that is, pack as many words as you can in each line. Pad extra spaces ' ' when necessary so that each line has exactly maxWidth characters.

Extra spaces between words should be distributed as evenly as possible. If the number of spaces on a line does not divide evenly between words, the empty slots on the left will be assigned more spaces than the slots on the right.

For the last line of text, it should be left-justified, and no extra space is inserted between words.

Note:

A word is defined as a character sequence consisting of non-space characters only.
Each word's length is guaranteed to be greater than 0 and not exceed maxWidth.
The input array words contains at least one word.


Example 1:

Input: words = ["This", "is", "an", "example", "of", "text", "justification."], maxWidth = 16
Output:
[
   "This    is    an",
   "example  of text",
   "justification.  "
]
Example 2:

Input: words = ["What","must","be","acknowledgment","shall","be"], maxWidth = 16
Output:
[
  "What   must   be",
  "acknowledgment  ",
  "shall be        "
]
Explanation: Note that the last line is "shall be    " instead of "shall     be", because the last line must be left-justified instead of fully-justified.
Note that the second line is also left-justified because it contains only one word.
Example 3:

Input: words = ["Science","is","what","we","understand","well","enough","to","explain","to","a","computer.","Art","is","everything","else","we","do"], maxWidth = 20
Output:
[
  "Science  is  what we",
  "understand      well",
  "enough to explain to",
  "a  computer.  Art is",
  "everything  else  we",
  "do                  "
]


Constraints:

1 <= words.length <= 300
1 <= words[i].length <= 20
words[i] consists of only English letters and symbols.
1 <= maxWidth <= 100
words[i].length <= maxWidth

```python
public class Solution {
    public IList<string> FullJustify(string[] words, int maxWidth) {

        int rowCount = 0;
        int capacity = 0;
        int spaces = 0;
        List<string> listPerRow = new List<string>();
        string[] output = new string[300*20+300];
        string textRow = "";
        for(int i =0; i<words.Length; i++){
            string word = words[i];

            capacity += word.Length;
            if(capacity+spaces>maxWidth){
                capacity -= word.Length;
                textRow = AlignJust(listPerRow, maxWidth, capacity);

                output[rowCount] = textRow;
                Console.WriteLine($"i: {i}. textRow: {textRow}.'");
                listPerRow.Clear();
                capacity = word.Length;
                spaces = 0;
                rowCount++;
                }
            spaces += 1;
            listPerRow.Add(word);
            //Console.WriteLine($"i: {i} rowCount: {rowCount}. capacity: {capacity}. word.Length: {word.Length}. word: {word}");
        }
        if(listPerRow.Count>0){
            textRow = AlignJust(listPerRow, maxWidth, capacity, true);
            }

            rowCount++;
            output[rowCount-1] = textRow;
            int spacesToFill = maxWidth-capacity;
            //Console.WriteLine($"spacesToFill: {spacesToFill}. textRow: {textRow} listPerRow: {listPerRow}");



        return output.Take(rowCount).ToArray();
    }

    public string AlignJust(List<string> listPerRow, int maxWidth, int capacity, bool lastLine = false) {
        string textRow = "";

        int spacesToFill = maxWidth-capacity;

        if(listPerRow.Count!=1){
            int gapPerWord = spacesToFill/(listPerRow.Count-1);
            int leftOver = spacesToFill%(listPerRow.Count-1);
            if(lastLine){
                string lastGap = new String(' ',maxWidth);
                listPerRow[listPerRow.Count-1]+=lastGap;
                Console.WriteLine($"lastLine spacesToFill: {spacesToFill}. gapPerWord: {gapPerWord}. leftOver: {leftOver}. lastGap: '{lastGap}'");
                gapPerWord=1;
                leftOver=0;

            }
            for(int j = 0; j<leftOver; j++ ){
                    listPerRow[j]+=" ";
            }

            string gaps = new String(' ',gapPerWord);
            //Console.WriteLine($"AlignJust List- spacesToFill: {spacesToFill}. gapPerWord: {gapPerWord}. gaps: '{gaps}'");
            textRow = string.Join(gaps,listPerRow);
        }
        else{
            string w = listPerRow.First();
            string g = new String(' ',maxWidth-w.Length);
            textRow = $"{w}{g}";
            Console.WriteLine($"AlignJust Single- w: {w}. g: {g}. textRow: '{textRow}'");
        }

        if(lastLine){
            textRow = textRow.Substring(0, maxWidth);
        }
        return textRow;
    }
}
```
